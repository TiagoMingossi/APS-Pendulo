```python
import math  
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


tempo = np.arange(0,5, 0.01)
g = 9.8
l = 2.5
m = 0.5


def Pendulo(A,tempo):
    ùúÉ = A[0]
    ùúî = A[1]
    dùúÉdt = ùúî
    dùúîdt = (-g*math.sin(ùúÉ))/l
    return [dùúÉdt, dùúîdt]


S0=[math.radians(90),0]

Solucao=odeint(Pendulo, S0, tempo)

eixo_x=[]
eixo_y=[]

for ùúÉ in Solucao[:,1]:
    eixo_x.append(l*math.cos(ùúÉ))
    
for ùúÉ in Solucao[:,0]:
    eixo_y.append(l*math.sin(ùúÉ))

plt.title('Trajetoria do Pendulo 90¬∞')   
plt.grid(True)
plt.plot(eixo_y,eixo_x)
plt.axis('equal')
plt.show()

#--------------------------------------------------------------

S1=[math.radians(45),0]

Solucao1=odeint(Pendulo, S1, tempo)

eixo_x1=[]
eixo_y1=[]

for ùúÉ in Solucao1[:,1]:
    eixo_x1.append(l*math.cos(ùúÉ))
    
for ùúÉ in Solucao1[:,0]:
    eixo_y1.append(l*math.sin(ùúÉ))

plt.title('Trajetoria do Pendulo 45¬∞')   
plt.grid(True)
plt.plot(eixo_y1,eixo_x1)
plt.axis('equal')
plt.show()

#--------------------------------------------------------------

S2=[math.radians(30),0]

Solucao2=odeint(Pendulo, S2, tempo)

eixo_x2=[]
eixo_y2=[]

for ùúÉ in Solucao2[:,1]:
    eixo_x2.append(l*math.cos(ùúÉ))
    
for ùúÉ in Solucao2[:,0]:
    eixo_y2.append(l*math.sin(ùúÉ))

plt.title('Trajetoria do Pendulo 30¬∞')   
plt.grid(True)
plt.plot(eixo_y2,eixo_x2)
plt.axis('equal')
plt.show()
```
